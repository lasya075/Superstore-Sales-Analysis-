```python
import numpy as np
import pandas as pd
```

```python
# reading data
df = pd.read_csv('train.csv')
```

```python
# checking the head of the data
df.head()
```

```python
# checking shape of the data"
df.shape
```

```python
# all the columns
df.columns
```

```python
# Replace spaces with underscores in column names
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

# Datatypes

```python
# summary of the data
df.info()
```

```python
df[["Order_Date","Ship_Date"]]
```

```python
#Changing the data type of 'Order_Date' and 'Ship_Date' to datetime type

df['Order_Date'] = pd.to_datetime(df['Order_Date'],format = "%d/%m/%Y")
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format = "%d/%m/%Y")

# checking if the format changed
df[["Order_Date","Ship_Date"]]
```

```python
df.info()
```

```python
#Converting the dtype of Postal_Code from float64 to int64:
df['Postal_Code'] = df['Postal_Code'].astype(str).str.strip()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

#Handling Duplicates

```python
#checking the duplicate records:

duplicate = df[df.iloc[:,1:].duplicated(keep=False)]
duplicate
```

```python
#Dropping the duplicate records:

cols_to_check = [col for col in df.columns if col != 'Row_ID']
df = df.drop_duplicates(subset=cols_to_check)
```

```python
df.shape
```

```python
#list of all column names:

cols = list(df.columns)

#Checking values count of each column:

for col in cols:
    print(col)
    print('')
    print(df[col].value_counts())
    print('')
```

```python
#Checking For each entry in dataset if ship date >= order date:

df[df['Ship_Date'] < df['Order_Date']]
```

```python
df.reset_index(drop = True,inplace = True)
```

```python
df
```

```python
df['Row_ID'] = range(1, len(df) + 1)
```

```python
df
```

```python
#Exporting the cleaned dataset to csv:

df.to_csv('superstore_sales.csv',index=False)
```
